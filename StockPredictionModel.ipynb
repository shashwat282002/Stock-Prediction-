```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import matplotlib as plt
from sklearn import preprocessing
```

```python
nift=pd.read_csv("/Users/shashwatbindal/Desktop/^NSEInew3.csv")
backup=nift

nift
```

```python
nift["High_prev"]=nift["High"]
nift["Low_curr"]=nift["Low"]
```

```python
nift["Open"]=nift.Open.shift(-1)
nift["Low_curr"]=nift.Low_curr.shift(-1)
nift["Close"]=nift.Close.shift(-1)
nift["High"]=nift.High.shift(-1)
nift
```

```python
nift=nift.dropna()
nift.tail()
```

```python
nift=nift.dropna()
nift
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor

x=nift[["Open","Low","High_prev","Adj Close"]]
y=nift["High"]
y_=nift["Low_curr"]
minv=99999
maxv=0
list1=[]
list2=[]
for i in range(50):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1)
    xtrain2,xtest2,ytrain2,ytest2=train_test_split(x,y_,random_state=1)

    model1=RandomForestRegressor()
    model2=RandomForestRegressor()

    #model=MLPRegressor(max_iter=500)


    model1.fit(xtrain.values,ytrain.values)
    model2.fit(xtrain2.values,ytrain2.values)

    data=pd.DataFrame([[18081.00,18042.40,18116.35,18089.85]])
    
    #data=pd.DataFrame([[4990.05,4953.35,4997.30,4985.85]])
    temp=model1.predict(data)
    temp2=model2.predict(data)
    list1.append(temp)
    list2.append(temp2)

#     if(minv>temp):
#         minv=temp
#     if(maxv<temp):
#         maxv=temp
# [[18326.099609,18246,18325.400391]]
ans1=np.median(list1)
ans2=np.median(list2)
print("Range is from ",ans2," to ",ans1)

#print("min value is: ",minv,"and max value is: ",maxv) 
```
